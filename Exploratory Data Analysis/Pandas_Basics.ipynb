#### 7. Handling Missing Values
* aka Data wrangling or Data Manipulations
* replace by MEAN if data is continous
* replace by MEDIAN if data is discrete / continous
* replace by MODE if data is categorical

```python
# replace is used to replace the specified value, inplace ensures the actual dataframe is updated 
raw_data.replace("?", np.nan, inplace = True)
```

```python
raw_data.isnull().sum() # Returns the total number of missing values from each columns names
```

```python
# Displaying the missing values in percentage
raw_data.isnull().sum() / raw_data.shape[0] * 100
```

```python
# Handling the missing values
# axis = 1 : Drops columns by using the specified column names
# axis = 0 : Drops rows by the specified labels
raw_data.drop(['symboling', 'normalized-losses'], axis = 1, inplace = True) # inplace applies the modification directly to the refernce DataFrame
```

```python
raw_data['price']
```

```python
raw_data['price'] = raw_data['price'].astype('float')
```

```python
raw_data['price']
```

```python
vehicle_df  = raw_data.copy()
vehicle_df
```

```python
vehicle_df['price'].median() # Median return the middle value of the data
```

```python
# replace the NAN values of price column with the median of the column
vehicle_df['price'] = vehicle_df['price'].fillna(vehicle_df['price'].median())
vehicle_df['price']
```

```python
# Drop missing rows where NAN values are present
vehicle_df.dropna(inplace = True)
vehicle_df.isnull().sum()
```

#### 8. Drop Duplicate Rows
* Drop duplicate values (if exactly all values of different rows are similar, then they are duplicate)

```python
# Returns True / False if any rows are duplicated or not
vehicle_df[vehicle_df.duplicated()]
```

```python
dict(vehicle_df.iloc[1])
```

```python
# Select random sample from dataframe using "frac" which takes value of sample in percentage or n = number of samples
sample_df = vehicle_df.sample(frac = 0.05)
sample_df
```

```python
vehicle_df1 = pd.concat([vehicle_df, sample_df], ignore_index = True)
vehicle_df1
```

```python
# Returns the duplicated rows
vehicle_df1[vehicle_df1.duplicated()]
```

```python
# Drops all the duplicated rows from the database
vehicle_df1.drop_duplicates(inplace=True)
vehicle_df1
```

#### 13. Merge

```python
data1 = {
    'EmployeeID': ['E001', 'E002', 'E003', 'E004', 'E005'],
    'Name': ['John Doe', 'Jane Smith', 'Peter Brown', 'Tom Johnson', 'Rita Patel'],
    'DeptID': ['D001', 'D003', 'D001', 'D002', 'D003']
}

data2 = {
    'DeptID': ['D001', 'D002', 'D003'],
    'DeptName': ['Sales', 'HR', 'Admin']
}
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

df1
```

```python
df2
```

```python
pd.merge(df1, df2, on='DeptID', how = 'inner') # It can do inner, outer, left, right
```
