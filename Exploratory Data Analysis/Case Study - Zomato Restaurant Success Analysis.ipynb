#### 6. Approx Cost Distribution for Two People

```python
zomato_df["approx_cost(for two people)"]
zomato_df.head()
zomato_df["approx_cost(for two people)"].isnull().sum()
zomato_df["approx_cost(for two people)"] = zomato_df["approx_cost(for two people)"].str.replace(",","").astype('float')
cost_for_two = zomato_df["approx_cost(for two people)"]
cost_for_two
```

```python
import warnings
warnings.filterwarnings('ignore')
sns.set_style('whitegrid')
sns.distplot(x = cost_for_two, color='red')
```

* The approx cost for two person is skewed towards positive. Hence it is Positive Skewness

```python
sns.boxplot(x = cost_for_two, color='red', orient='h')
```

```python
zomato_df["approx_cost(for two people)"].median()
```

```python
Q1 = zomato_df["approx_cost(for two people)"].quantile(0.25) # First Quartile
Q2 = zomato_df["approx_cost(for two people)"].quantile(0.5) # Median Quartile
Q3 = zomato_df["approx_cost(for two people)"].quantile(0.75) # Third Quartile
print(f"First Quartile: {Q1}, Median Quartile: {Q2}, Third Quartile: {Q3}")
```

```python
IQR = Q3 - Q1 # Inter-quartile range
print(f"Inter-Quartile Range of Cost for two People: {IQR}")
```

* The approx cost for two person ranges from 300 to 650 INR. The approx cost for two person is skewed towards positive. Hence it is Positive Skewness
