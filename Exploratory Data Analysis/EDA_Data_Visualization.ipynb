## Univariate Analysis
* Univariate analysis is the simplest form of analysis
* Only one variable is used in visualization
* It doesn't deal with relationship plots. It takes data, summarize the data, pattern analysis, trend analysis in data.
* Univariate analysis can be implemented on **numerical data** or **categorical data**

```python
# Selects attributes with the specified data type
df_numerical = housing_data.select_dtypes(include = ['number'])
df_numerical
```

```python
df_categorical = housing_data.select_dtypes(include = ['O'])
df_categorical.info()
```

### 1. Line Plot
* Univariate (single variable) - Trend Analysis
* Bivariate (two variable) - Relationship Plot
* Implemented for Trend Analysis and Trend Behaviour

```python
# Inbuilt Pandas Function to Plot Graphs using - plot()
df_numerical['SalePrice'].plot(kind = 'line')
```

```python
plt.figure(figsize = [5, 5])
plt.style.use('ggplot')
plt.plot(df_numerical['SalePrice'].sort_values(ascending = True).unique(), color = 'orangered')
plt.xlabel("Index")
plt.ylabel("Sales Price")
plt.title("Sales Price Trends")
plt.show()
```

### 2. Bar Plot
* Univariate or Bivariate Plot
* Bar Chart are used with categorical dataset
* They are used to display counts of uique values for categorical type
* Heigh of bars indicate the count for each category

```python
df_categorical
```

```python
df_categorical['MoSold'].value_counts().values
plt.figure(figsize = [7, 8])
plt.xlabel("Month")
plt.ylabel("Sales per Month")
plt.title("Monthly Sales Price Trends", size = 12)
plt.bar(df_categorical['MoSold'].value_counts().index, height = df_categorical['MoSold'].value_counts().values)
```

```python
df_categorical['MoSold'].value_counts().plot(kind = 'bar', color = 'green', figsize = [4, 3])
```

### 3. Histogram
* Univariate Analysis - Apply on continous data.
* Frequency of Data in range
* Study of Spread / Data Distribution

```python
df_numerical
```

```python
# bins : it is used to specify the distribution of the data range between which the frequency will be counted
plt.style.use('ggplot')
plt.figure(figsize=[6, 5])
plt.hist(df_numerical['SalePrice'], bins = 30) 
```

### 5. Scatter Plot
* Indicates relationship plot or correlation between two numerical columns

```python
plt.figure(figsize = [5, 4])
plt.scatter(df_numerical[df_numerical['LotArea'] < 60000]['LotArea'], 
            df_numerical[df_numerical['LotArea'] < 60000]['SalePrice'])
plt.xlabel("Lot Area")
plt.ylabel("Sales Price")
plt.title("Scatter Plot - Relationship b/w LotArea and SalePrice", size = 10)
plt.show()
```

### 6. Seaborn Plot

```python
sns.scatterplot(x = df_numerical[df_numerical['LotArea'] < 60000]['LotArea'], 
            y =df_numerical[df_numerical['LotArea'] < 60000]['SalePrice'],
            hue = df_categorical['GarageType'], s = 50)
plt.show()
```

```python
sns.scatterplot(x = df_numerical['GarageArea'], 
            y =df_numerical['SalePrice'],
            hue = df_categorical['GarageType'], s = 50, marker='*')
# s = size of bubble
# marker = type of bubble ('s', '^', '>', '<', '*', 'd', 'o', '+', , '-')
# alpha = transparency of each bubble
plt.show()
```

### 7. Pie-Chart

### 8. Distribution Plot / Kernel Density Estimator (KDE) Plot
* Shows historgram + density plot to display distribution of numerical continous data.

```python
import warnings
warnings.filterwarnings('ignore')
sns.distplot(df_numerical[df_numerical['GarageArea'] > 0]['GarageArea'], hist=False, color = 'g')
plt.show()
```

### 9. Box Plot
* Box Plot is dividided into a two parts.
* It contains 5 important aspects : min val, Q1(25%), Q2(50%), Q3(75%), Max val
* min val = Q1 - 1.5 x IQR
* max val = Q3 + 1.5 x IQR (Inter Quartile Range)
* 

```python
plt.figure(figsize=[10,4])
sns.boxplot(df_numerical[df_numerical['GarageArea']>0]['GarageArea'], orient = 'h')
plt.show()
```
